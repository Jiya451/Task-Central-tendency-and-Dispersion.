# **Dealing with outliers**

Using Quartiles to remove outliers.

```python
    # Calculate quartiles
    Q1 = df['BMXWAIST'].quantile(0.25)
    Q3 = df['BMXWAIST'].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1
    print(Q1)
    print(Q3)
    print(IQR)
```

```python
lower_bound= Q1-1.5* IQR
print(lower_bound)
upper_bound= Q3+1.5* IQR
print(upper_bound)
```

```python
outliers = df[(df['BMXWAIST'] < lower_bound) | (df['BMXWAIST'] > upper_bound)]['BMXWAIST']
#outliers = df[(df['BMXWAIST'] > lower_bound) & (df['BMXWAIST'] < upper_bound)]
print("These values are outliers:")
print(outliers)
print("Number of outliers:", outliers.count())
```

```python
plt.figure(figsize = (20 , 5 ))
#actual data and outliers
plt.scatter(df['BMXWAIST'].index, df['BMXWAIST'] , color='green', label='Data points' , s=0.5)
plt.scatter(outliers.index, outliers, color='red', label='Outliers',s=0.5)

plt.title('Outliers in BMXWAIST')
plt.xlabel('ID')
plt.ylabel('WAIST')
plt.legend()
plt.show()
```

```python
# Deleting the outliers
df = df[(df['BMXWAIST'] >= lower_bound) & (df['BMXWAIST'] <= upper_bound)]
```

```python
# Calculating mean after removing outliers
mean_without_outliers = df['BMXWAIST'].mean()

print("Mean without outliers:", mean_without_outliers)
```
